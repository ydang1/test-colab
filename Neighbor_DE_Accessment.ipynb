```python
## predicted LTV we already in the listing_predictions table the prediction date is the same as listed_date
## need to get the acquisition cost to user's join_date

##date need

## fact table

# fct_cohort_date_LTV
# listing_id
# user_id
# join_date --> user_join_date
# join_date_year
# join_date_month
# join_date_week
# listing_date
# predicted_ltv
# cac

fct_cohort_date_ltv = pd.read_sql("""
WITH listings AS (
SELECT listing_id, host_user_id, listed_date
FROM listings_daily_snapshot
GROUP BY 1,2,3
),

users AS (
SELECT user_id, join_date, acquisition_channel
FROM users_daily_snapshot
GROUP BY 1,2,3
),

pred_ltv AS (
SELECT listing_id, host_ltv, predicted_at
FROM listing_predictions
)

SELECT
A.user_id, A.join_date, A.acquisition_channel, B.listing_id, B.listed_date, C.host_ltv,
strftime('%Y-01-01', A.join_date) AS join_date_year,
strftime('%Y-%m-01', A.join_date) AS join_date_month,
date(A.join_date, 'weekday 0', '-6 days') AS join_date_week
FROM users A
LEFT JOIN listings B
ON A.user_id = B.host_user_id
LEFT JOIN pred_ltv C
ON B.listing_id = C.listing_id
AND DATE(B.listed_date) = Date(C.predicted_at)
"""
, con)


fct_cohort_date_ltv.to_sql("fct_cohort_date_ltv", con, if_exists="replace", index=False)


```

```python
pd.read_sql("SELECT * FROM fct_cohort_date_ltv WHERE listing_id IS NOT NULL LIMIT 5", con)
```

```python
## fact table for CAC

## issues: we can't perfectly match each dolloar to each user, we distribute the spent evenly across attributed signups:
## ideally what we want to attribute cost to the day users joined per user level but it is hard since we do not have grainularity for ads costs per each user
## alternatively we calculate the CAC by join Date, since we can't perfectly match each dollar to each user, we distribute the spend evenly across attribted signups;

## user_id
## join_date
## acquisition_channel
## total_spend
## total_attributed_signups
## cac as attributed_cac

fct_cohort_date_cac= pd.read_sql(
    """
    WITH users AS (
    SELECT user_id, join_date, acquisition_channel
    FROM users_daily_snapshot
    GROUP BY 1,2,3
    ),
    spend_by_channel_date AS (
    SELECT date, channel, SUM(spend_amount) AS total_spend,
    SUM(attributed_signups) AS total_attributed_signups,
    SUM(spend_amount) / NULLIF(SUM(attributed_signups), 0) AS cac
    FROM ad_spend_daily
    GROUP BY 1,2
    )

    SELECT A.user_id, A.join_date, A.acquisition_channel, B.total_spend, B.total_attributed_signups, B.cac
    FROM users A LEFT JOIN spend_by_channel_date B
    ON A.join_date = B.date
    AND A.acquisition_channel = B.channel
    """, con)


fct_cohort_date_cac.to_sql("fct_cohort_date_cac", con, if_exists="replace", index=False)
```

```python
##final step calculated the final LTV:CAC value
## daily result
pd.read_sql (
  """
  SELECT A.join_date, SUM(COALESCE(A.host_ltv, 0)) AS host_ltv,
  SUM(COALESCE(B.cac,0)) AS cac,
  SUM(COALESCE(A.host_ltv, 0)) / NULLIF(SUM(COALESCE(B.cac,0)), 0) AS ltv_cac
  FROM fct_cohort_date_ltv A INNER JOIN fct_cohort_date_cac B
  ON A.user_id = B.user_id
  AND A.join_date = B.join_date
  AND A.acquisition_channel = B.acquisition_channel
  --AND (A.acquisition_channel IS NULL OR A.acquisition_channel = 'organic')
  GROUP BY 1
  --HAVING SUM(COALESCE(A.host_ltv, 0)) = 0
  """, con
)
```

```python
## yearly results
##final step calculated the final LTV:CAC value
## daily result
pd.read_sql (
  """
  SELECT A.join_date_year, SUM(COALESCE(A.host_ltv, 0)) AS host_ltv,
  SUM(COALESCE(B.cac,0)) AS cac,
  SUM(COALESCE(A.host_ltv, 0)) / NULLIF(SUM(COALESCE(B.cac,0)), 0) AS ltv_cac
  FROM fct_cohort_date_ltv A INNER JOIN fct_cohort_date_cac B
  ON A.user_id = B.user_id
  AND A.join_date = B.join_date
  AND A.acquisition_channel = B.acquisition_channel
  GROUP BY 1
  --HAVING SUM(COALESCE(A.host_ltv, 0)) / NULLIF(SUM(COALESCE(B.cac,0)), 0) < 1 AND SUM(COALESCE(A.host_ltv, 0)) / NULLIF(SUM(COALESCE(B.cac,0)), 0) > 0
  ORDER BY 1
  """, con
)
```
